### Grey Level Transforms

```python
from PIL import Image
import numpy as np
from numpy import *
from matplotlib import pylab
from pylab import *

image = array(Image.open(r"C:\Users\Wavicledata\Pictures\Saved Pictures\mini-cooper.jpg").convert("L"))
```

```python
imshow(image)
```

```python
## Invert Image ##
image_2 = 255 - image
imshow(image_2)
```

```python
## Clamp to Interval 100 to 200 ##
image_3 = (100.0/255) * image + 100
imshow(image_3)
```

```python
## Squarred Image ##
image_4 = 255.0 * (image /255.0) ** 2
imshow(image_4)
```

```python
print(int(image.min()), int(image.max()))
```

```python
## Reverse of the array transformation can be done in PIL ##
pil_im = Image.fromarray(image)
imshow(pil_im)
```
